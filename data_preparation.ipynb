```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error
from sklearn.metrics.pairwise import cosine_similarity
import dill as pickle
```

```python
# Load dataset
df = pd.read_csv('school_dataset.csv')
df.head()
```

```python
df['performance'] = df[['performance_2023', 'performance_2022', 'performance_2021']].mean(axis='columns') 

df.head()
```

```python

# Label encoding for categorical features
label_encoders = {}
categorical_features = ['religion', 'gender', 'boarding/day']
for feature in categorical_features:
    le = LabelEncoder()
    df[feature] = le.fit_transform(df[feature])
    label_encoders[feature] = le
```

```python

# Standardizing numerical features
scaler = StandardScaler()
numerical_features = ['performance', 'Latitude', 'Longitude', 'stars']
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.1, random_state=42)

# Creating feature matrix excluding the 'School' column
train_feature_matrix = train_df.drop(columns=['School'])
test_feature_matrix = test_df.drop(columns=['School'])

# Train the k-NN model using cosine similarity
knn_model = NearestNeighbors(n_neighbors=10, metric='cosine')
knn_model.fit(train_feature_matrix)
```

```python

# Preprocessing function for user preferences
def preprocess_user_preferences(user_preferences, label_encoders, scaler):
    # Encode categorical features
    for feature in ['religion', 'gender', 'boarding/day']:
        if user_preferences[feature] not in label_encoders[feature].classes_:
            # Handle unseen labels by adding them to the encoder
            label_encoders[feature].classes_ = np.append(label_encoders[feature].classes_, user_preferences[feature])
        user_preferences[feature] = label_encoders[feature].transform([user_preferences[feature]])[0]
    
    # Creating DataFrame for the user preferences
    user_df = pd.DataFrame([user_preferences])
    
    # Removing 'location' column before scaling
    user_df.drop(columns=['location'], inplace=True)
    
    # Standardize numerical features
    user_df[['performance', 'Latitude', 'Longitude', 'stars']] = scaler.transform(user_df[['performance', 'Latitude', 'Longitude', 'stars']])
    
    return user_df
```

```python

# Similarity calculation function (cosine similarity)
def calculate_similarity(user_df, feature_matrix):
    extended_feature_matrix = pd.concat([feature_matrix, user_df], ignore_index=True)
    cosine_sim = cosine_similarity(extended_feature_matrix)
    user_sim_scores = cosine_sim[-1, :-1]
    return user_sim_scores
```

```python

# Precision at k evaluation
def precision_at_k(test_df, knn_model, train_df, k=10, performance_threshold=2.5):
    precision_scores = []

    for _, row in test_df.iterrows():
        user_preferences = row.drop(labels=['School']).to_dict()
        user_preferences['location'] = 'test_location'  # Dummy location

        user_df = preprocess_user_preferences(user_preferences, label_encoders, scaler)
        distances, indices = knn_model.kneighbors(user_df)
        recommended_indices = indices[0]
        recommended_schools = train_df.iloc[recommended_indices]

        # Determine if the top k recommendations are relevant
        relevant_recommendations = recommended_schools[recommended_schools['performance'] <= performance_threshold]
        precision = len(relevant_recommendations) / k
        precision_scores.append(precision)

    avg_precision = np.mean(precision_scores)
    return avg_precision

# Calculating Precision at k
p_at_k = precision_at_k(test_df, knn_model, train_df, k=10, performance_threshold=2.5)
print(f"Precision at 10: {p_at_k}")
```
